```python
import yaml
import pandas as pd
%matplotlib inline
```

```python
def plot(filename):
    with open(filename, 'r') as f:
        hist = yaml.load(f)
    for model, history in hist.items():
        h = pd.DataFrame(history)
        h.columns = pd.MultiIndex.from_tuples(h.columns)
        for metric in h.columns.levels[0]:
            h[metric][2:].plot(title='%s: %s'%(model, metric))
```

```python
plot('model/test.hist')
```
